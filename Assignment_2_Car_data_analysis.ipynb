```python
#import pandas library
import pandas as pd
df = pd.read_csv("car data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.drop ( columns = ['experience'])
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(df['speed'], df['car_age'])
plt.xlabel("speed")
plt.ylabel("risk")
plt.title("speed vs risk")
plt.show()
```

```python
x_mean = df['speed'].mean()
y_mean = df['car_age'].mean()
```

```python
plt.scatter(x_mean, y_mean, color = "red")
plt.scatter(df['speed'], df['car_age'])
plt.xlabel("speed")
plt.ylabel("risk")
plt.title("speed vs risk")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(df[['speed']], df['car_age'])
```

```python
reg.coef_
reg.intercept_
```

```python
reg.score(df[['speed']], df['car_age'])
```

```python
plt.plot(df['speed'], reg.predict(df[['speed']]))
plt.scatter(df['speed'], df['car_age'])
plt.xlabel("speed")
plt.ylabel("risk")
plt.title("speed vs risk")
plt.show()
```
