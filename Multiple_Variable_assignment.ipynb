```python
#importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Create data frame
df = pd.read_csv("car data.csv")
df.head()
```

```python
#Find any missing data
df.isnull().sum()
```

```python
#ploting graph
plt.scatter(df['speed'],df['risk'], color = "blue")
plt.scatter(df['car_age'],df['risk'], color = "red")
plt.scatter(df['experience'],df['risk'], color = "green")
plt.xlabel('speed')
plt.ylabel('risk')
plt.title('Risk Analysis')
plt.show()
```

```python
#linear Regression model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(df[['speed','car_age','experience']],df.risk)
```

```python
#Predict risk
reg.predict([[150,10,23]])
```

```python
#find r2 value
reg.score(df[['speed','car_age','experience']],df.risk)
```

```python
reg.coef_
reg.intercept_
```
