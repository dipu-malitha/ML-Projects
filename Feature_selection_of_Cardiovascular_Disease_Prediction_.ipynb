```python
import pandas as pd
df = pd.read_csv('cardio_train.csv', sep = ";")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
x = df.drop("cardio",axis =1)
y = df["cardio"]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
feature = SelectKBest(score_func=f_classif,k="all")
fit = feature.fit(x,y)
```

```python
score_col = pd.DataFrame(fit.scores_, columns=["Score Value"])
name_col = pd.DataFrame(x.columns, columns= [" Name"])
```

```python
score_col
```

```python
final_feature = pd.concat([name_col,score_col],axis=1)
final_feature
```

```python
final_feature.sort_values("Score Value",ascending=False)
```

```python
final_feature.nlargest(6,"Score Value")
```

```python
x = df[["age","ap_hi","ap_lo","cholesterol","gluc","weight"]]
y = df["cardio"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
