```python
#import pandas and matplotlib library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the csv file and create data file
df = pd.read_csv('marital status.csv')
df.head()
```

```python
#find any null cell
df.isnull().sum()
```

```python
#fill up the null cell
handle = df['status'].median()
df.status = df.status.fillna(handle)
df.head()
```

```python
#Check again any null file
df.isnull().sum()
```

```python
#ploting the data set
plt.scatter(df.age,df.status)
plt.xlabel('age')
plt.ylabel('status')
plt.title('Age vs material status')
plt.show()
```

```python
#find x mean and y mean
x_mean = df.age.mean()
y_mean = df.status.mean()
x_mean
y_mean
```

```python
#ploting mean
plt.scatter(df.age,df.status)
plt.xlabel('age')
plt.ylabel('status')
plt.title('Age vs material status')
plt.plot([x_mean,x_mean],[0,y_mean],color='red')
```

```python
#import linear regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
#fit the line
reg.fit(df[['age']],df.status)
```

```python
#predict value
reg.predict([[40]])
```

```python
#Add predict value in the data frame
df['predict'] = reg.predict([[i] for i in df.age])
df.head()
```

```python
#score
reg.score(df[['age']],df.status)
```
