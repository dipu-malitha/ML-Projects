```python
#Import libray for data frame and ploting
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the csv file
df = pd.read_csv('Salary Data.csv')
df.head()
```

```python
#Check have any null data
df.isnull().sum()
```

```python
#Ploting data
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

Linear regression

```python
#import linear regression from sklearn library
from sklearn.linear_model import LinearRegression
```

```python
#Assign linear regression
reg = LinearRegression()
```

```python
# Find x axis mean and y axis mean
x = df['Experience Years']
y = df['Salary']
x.mean()
y.mean()
```

```python
#Ploting the regression graph
plt.scatter(x.mean(),y.mean(), color = "red")
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

```python
#estimate the regression
x = df[['Experience Years']]
y = df[['Salary']]
reg.fit(x, y)
```

```python
#find co-efficient and intercept
print(reg.coef_)
print(reg.intercept_)
```

```python
#predict a value
reg.predict([[5]])
```

```python
#Include predict value in Data frame
df['Salary_predict'] = reg.predict(df[['Experience Years']])
df.head()
```

```python
#Plot predicted value with other parameters
plt.scatter(df['Experience Years'], df['Salary'])
plt.plot(df['Experience Years'], df['Salary_predict'], color = 'red')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
```

Loss and cost function

```python
#Calculate loss value
df["lose"] = df["Salary"] - df["Salary_predict"]
df.head()
```

```python
#Find MSE and MAE value
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(df["Salary"], df["Salary_predict"])
mae = mean_absolute_error(df["Salary"], df["Salary_predict"])
print(mse)
print(mae)
```

R2 value or accuracy for regression

```python
#Regression value or score
reg.score(x, y)
```

```python
from sklearn.metrics import r2_score
r2_score(df["Salary"], df["Salary_predict"])
```
